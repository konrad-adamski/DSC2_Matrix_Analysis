# MDS Zusammenf√ºhrung

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dsc_path = "../data/010_dsc.csv"
```

## 1) Mit Mean

```python
# Einlesen der CSV-Datei
df_experiments_mds = pd.read_csv("data/mds_intervals_mean.csv", index_col=0)
df_experiments_mds
```

```python
# Visualisierung der MDS-Ergebnisse
plt.figure(figsize=(10, 8))
plt.scatter(df_experiments_mds["MDS1_mean"], df_experiments_mds["MDS2_mean"], c='blue', alpha=0.6)
plt.title("Relative Position der Proben (MDS)")
plt.xlabel("MDS1_mean")
plt.ylabel("MDS2_mean")

# Beschriftung der Punkte mit den Probenbezeichnungen
for i, txt in enumerate(df_experiments_mds.index):
    plt.annotate(txt, (df_experiments_mds["MDS1_mean"].iloc[i], df_experiments_mds["MDS2_mean"].iloc[i]))

plt.show()
```

```python
df_dsc = pd.read_csv(dsc_path, index_col=0)
df_dsc
```

```python
merged_df = df_dsc.join(df_experiments_mds, how="left")
merged_df
```

```python
# Manuelle Zuordnung der Farben zu den "Matrix"-Kategorien
category_color_map = {
    "Modellsand": "#FFD580",  # Helles Orange
    "Sand": "#FF7F00",        # Dunkleres Orange
    "Sediment Wei√üeritz": "green",
    "Sediment Elbe": "red"
}
```

```python
# Unique categories in the "Matrix" column
unique_categories = merged_df["Matrix"].unique()

# Map the colors to the "Matrix" column
merged_df["Color_Code"] = merged_df["Matrix"].map(category_color_map)

# Plotting
plt.figure(figsize=(10, 8))
for category, color in category_color_map.items():
    subset = merged_df[merged_df["Matrix"] == category]
    plt.scatter(subset["MDS1_mean"], subset["MDS2_mean"], label=category, color=color, alpha=0.6)

# Add legend
plt.legend(title="Matrix Kategorien", loc='best')
plt.title("Relative Position der Proben (MDS)")
plt.xlabel("MDS1_mean")
plt.ylabel("MDS2_mean")
plt.grid(True)
plt.show()

```

## 2) Pro Intervall

```python
# Einlesen der CSV-Datei
df_experiments_mds = pd.read_csv("data/mds_intervals.csv", index_col=0)
df_experiments_mds
```

```python
merged_df = df_dsc.join(df_experiments_mds, how="left")
merged_df
```

```python
import matplotlib.pyplot as plt

# Identifizieren der Intervalle aus den MDS-Spalten
mds_intervals = [
    col.split("_")[-1] for col in merged_df.columns if col.startswith("MDS1_")
]

# Unique categories in the "Matrix" column
unique_categories = merged_df["Matrix"].unique()

# Iterative Plots f√ºr jedes Intervall
for interval in mds_intervals:
    mds1_col = f"MDS1_{interval}"
    mds2_col = f"MDS2_{interval}"

    # Plot f√ºr das aktuelle Intervall erstellen
    plt.figure(figsize=(10, 8))
    
    # Punkte nach Kategorie plotten
    for category, color in category_color_map.items():
        subset = merged_df[merged_df["Matrix"] == category]
        plt.scatter(
            subset[mds1_col], 
            subset[mds2_col], 
            label=category, 
            color=color, 
            alpha=0.6
        )
    
    # Titel und Achsenbeschriftung
    plt.title(f"Relative Position der Proben (MDS):  {interval}")
    plt.xlabel("MDS1")
    plt.ylabel("MDS2")
    plt.legend(title="Matrix Kategorien", loc='best')
    plt.grid(True)
    plt.show()
```
