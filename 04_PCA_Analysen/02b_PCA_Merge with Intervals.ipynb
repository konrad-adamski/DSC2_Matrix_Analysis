# PCA Zusammenf√ºhrung

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dsc_path = "../data/010_dsc.csv"
```

## 1) Mit Mean

```python
# Einlesen der CSV-Datei
df_experiments_pca = pd.read_csv("data/pca_intervals_mean.csv", index_col=0)
df_experiments_pca
```

```python
# Visualisierung der PCA-Ergebnisse
plt.figure(figsize=(10, 8))
plt.scatter(df_experiments_pca["PCA1_mean"], df_experiments_pca["PCA2_mean"], c='blue', alpha=0.6)
plt.title("Relative Position der Proben (PCA)")
plt.xlabel("PCA1")
plt.ylabel("PCA2")

# Beschriftung der Punkte mit den Probenbezeichnungen
for i, txt in enumerate(df_experiments_pca.index):
    plt.annotate(txt, (df_experiments_pca["PCA1_mean"].iloc[i], df_experiments_pca["PCA2_mean"].iloc[i]))

plt.show()
```

```python
df_dsc = pd.read_csv(dsc_path, index_col=0)
df_dsc
```

```python
merged_df = df_dsc.join(df_experiments_pca, how="left")
merged_df
```

```python
# Manuelle Zuordnung der Farben zu den "Matrix"-Kategorien
category_color_map = {
    "Modellsand": "#FFD580",  # Helles Orange
    "Sand": "#FF7F00",        # Dunkleres Orange
    "Sediment Wei√üeritz": "green",
    "Sediment Elbe": "red"
}

# Map the colors to the "Matrix" column
merged_df["Color_Code"] = merged_df["Matrix"].map(category_color_map)

# Plotting
plt.figure(figsize=(10, 8))
for category, color in category_color_map.items():
    subset = merged_df[merged_df["Matrix"] == category]
    plt.scatter(subset["PCA1_mean"], subset["PCA2_mean"], label=category, color=color, alpha=0.6)

# Add legend
plt.legend(title="Matrix Kategorien", loc='best')
plt.title("Relative Position der Proben (PCA)")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.grid(True)
plt.show()
```

## 2) Pro Intervall

```python
# Einlesen der CSV-Datei
df_experiments_pca = pd.read_csv("data/pca_intervals.csv", index_col=0)
df_experiments_pca
```

```python
# Zusammenf√ºhren der Daten
merged_df = df_dsc.join(df_experiments_pca, how="left")
merged_df
```

```python
import matplotlib.pyplot as plt

# Manuelle Zuordnung der Farben zu den "Matrix"-Kategorien
category_color_map = {
    "Modellsand": "#FFD580",  # Helles Orange
    "Sand": "#FF7F00",        # Dunkleres Orange
    "Sediment Wei√üeritz": "green",
    "Sediment Elbe": "red"
}


# Identifizieren der Intervalle aus den PCA-Spalten
pca_intervals = [
    col.split("_")[-1] for col in merged_df.columns if col.startswith("PCA1_")
]

# Iterative Plots f√ºr jedes Intervall
for interval in pca_intervals:
    pca1_col = f"PCA1_{interval}"
    pca2_col = f"PCA2_{interval}"

    # Plot f√ºr das aktuelle Intervall erstellen
    plt.figure(figsize=(10, 8))
    
    # Punkte nach Kategorie plotten
    for category, color in category_color_map.items():
        subset = merged_df[merged_df["Matrix"] == category]
        plt.scatter(
            subset[pca1_col], 
            subset[pca2_col], 
            label=category, 
            color=color, 
            alpha=0.6
        )
    
    # Titel und Achsenbeschriftung
    plt.title(f"Relative Position der Proben (PCA): {interval}")
    plt.xlabel("PCA1")
    plt.ylabel("PCA2")
    plt.legend(title="Matrix Kategorien", loc='best')
    plt.grid(True)
    plt.show()
```
