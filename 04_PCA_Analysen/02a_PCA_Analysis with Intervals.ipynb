# Hauptkomponentenanalyse mit euklidischer Distanz

## 1) Einlesen der Daten

```python
file_path = "../data/030_dsc_experiments.csv"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Einlesen der CSV-Datei
df_experiments = pd.read_csv(file_path, index_col=0)
df_experiments
```

## 2) Filtern der relevanten Daten (Segment "_S5")

```python
# Filtern der Spalten, die "_S5" enthalten (Segment 5 "Erhitzen")
columns_s5 = [col for col in df_experiments.columns if "_S5" in col]
df_heating = df_experiments[columns_s5]
df_heating
```

## 3) Normalisierung der Daten

```python
from sklearn.preprocessing import StandardScaler

# Normalisierung der Daten
scaler = StandardScaler()
data_normalized = scaler.fit_transform(df_heating)

# Umwandlung zurück in ein DataFrame
df_heating_normalized = pd.DataFrame(data_normalized, columns=df_heating.columns, index=df_heating.index)
df_heating_normalized
```

```python
# Intervallgrenzen von -20 bis 270 in 20-Schritten
interval_groups = list(range(-20, 271, 20))

# Dictionary aus DataFrames für jedes Intervall erstellen
interval_dataframes = {
    f"{interval_groups[i]} bis {interval_groups[i + 1]}": 
    df_heating_normalized.loc[
        (df_heating_normalized.index >= interval_groups[i]) & 
        (df_heating_normalized.index < interval_groups[i + 1])
    ]
    for i in range(len(interval_groups) - 1)
}
```

```python
# Beispiel: Zugriff auf den DataFrame für ein Intervall
example_interval_df = interval_dataframes["-20 bis 0"]  # Ersetze mit einem gewünschten Intervall

# Ergebnis anzeigen
example_interval_df
```

## 4) Transposition der Intervalle

```python
# Alle DataFrames im Dictionary transponieren
transposed_interval_dataframes = {
    key: df.T for key, df in interval_dataframes.items()
}

# Beispiel: Zugriff auf den transponierten DataFrame für ein Intervall
example_transposed_interval_df = transposed_interval_dataframes["-20 bis 0"]  # Ersetze mit einem gewünschten Intervall

# Ergebnis anzeigen
example_transposed_interval_df.head()
```

```python
# Transposition der Datenmatrix: Proben werden Zeilen
df_basis = df_heating_normalized.T
df_basis
```

## 5) Reduktion der Dimension mit PCA

```python
from sklearn.decomposition import PCA

# Iterieren über alle Intervalle
for interval, df in transposed_interval_dataframes.items():
    # PCA auf das transponierte DataFrame anwenden
    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(df)

    # PCA-Ergebnisse zu df_basis hinzufügen
    df_basis[f"PCA1_{interval}"] = pca_result[:, 0]
    df_basis[f"PCA2_{interval}"] = pca_result[:, 1]

# Ergebnis: df_basis enthält jetzt die PCA-Komponenten für alle Intervalle
df_basis
```

## 7) Visualisierung der PCA-Ergebnisse

```python
import matplotlib.pyplot as plt

# Vorbereitung für den Plot
plt.figure(figsize=(12, 8))

# Alle PCA-Spalten finden
pca_columns = [col for col in df_basis.columns if isinstance(col, str) and col.startswith("PCA1_")]

# Iteriere über die PCA-Komponenten für jedes Intervall
for pca1_col in pca_columns:
    # Den zugehörigen PCA2-Spaltennamen ermitteln
    pca2_col = pca1_col.replace("PCA1_", "PCA2_")
    
    # Extrahiere das Intervall aus dem Spaltennamen
    interval = pca1_col.replace("PCA1_", "")
    
    # Streudiagramm erstellen
    plt.scatter(
        df_basis[pca1_col], 
        df_basis[pca2_col], 
        label=interval, 
        alpha=0.6
    )

# Achsenbeschriftung und Titel
plt.title("Visualisierung der PCA-Ergebnisse für alle Intervalle")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.legend(title="Intervalle")
plt.grid(True)
plt.show()

```

## 8) Export als CSV

```python
df_basis.index = df_basis.index.astype(str).str.replace('_S5', '', regex=False)
df_basis
```

```python
df_basis.to_csv("data/pca_intervals.csv", index=True)
```

## 9) Mitteln

```python
# Mittelung der PCA-Werte über alle Intervalle
pca1_columns = [col for col in df_basis.columns if isinstance(col, str) and col.startswith("PCA1_")]
pca2_columns = [col for col in df_basis.columns if isinstance(col, str) and col.startswith("PCA2_")]

# Berechnung des Durchschnitts für alle Intervalle
df_basis["PCA1_mean"] = df_basis[pca1_columns].mean(axis=1)
df_basis["PCA2_mean"] = df_basis[pca2_columns].mean(axis=1)

# Scatterplot mit zusammengeführten PCA-Werten
plt.figure(figsize=(10, 8))
plt.scatter(df_basis["PCA1_mean"], df_basis["PCA2_mean"], alpha=0.7, color="blue")

# Achsenbeschriftung und Titel
plt.title("Zusammengeführte PCA-Ergebnisse (ein Punkt pro Probe)")
plt.xlabel("PCA1 (gemittelt)")
plt.ylabel("PCA2 (gemittelt)")
plt.grid(True)
plt.show()

```
