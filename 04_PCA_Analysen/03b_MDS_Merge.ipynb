# MDS Zusammenführung

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dsc_path = "../data/010_dsc.csv"
```

## 1) Einlesen der Daten

```python
# Einlesen der CSV-Datei
df_experiments_mds = pd.read_csv("data/mds_simple.csv", index_col=0)
df_experiments_mds
```

```python
# Visualisierung der MDS-Ergebnisse
plt.figure(figsize=(10, 8))
plt.scatter(df_experiments_mds["MDS1"], df_experiments_mds["MDS2"], c='blue', alpha=0.6, edgecolor='k', s=70)
plt.title("Relative Position der Proben (MDS)", fontsize=14)
plt.xlabel("MDS1", fontsize=12)
plt.ylabel("MDS2", fontsize=12)

# Beschriftung der Punkte mit den Probenbezeichnungen
for i, txt in enumerate(df_experiments_mds.index):
    plt.annotate(txt, (df_experiments_mds["MDS1"].iloc[i], df_experiments_mds["MDS2"].iloc[i]),
                 fontsize=6, textcoords="offset points", xytext=(5, 5))

plt.grid(True, alpha=0.3)
plt.show()
```

```python
df_dsc = pd.read_csv(dsc_path, index_col=0)
df_dsc
```

```python
merged_df = df_dsc.join(df_experiments_mds, how="left")
merged_df
```

```python
# Manuelle Zuordnung der Farben zu den "Matrix"-Kategorien
category_color_map = {
    "Modellsand": "#FFD580",  # Helles Orange
    "Sand": "#FF7F00",        # Dunkleres Orange
    "Sediment Weißeritz": "green",
    "Sediment Elbe": "red"
}
```

```python
# Unique categories in the "Matrix" column
unique_categories = merged_df["Matrix"].unique()


# Map the colors to the "Matrix" column
merged_df["Color_Code"] = merged_df["Matrix"].map(category_color_map)

# Plotting
plt.figure(figsize=(10, 8))
for category, color in category_color_map.items():
    subset = merged_df[merged_df["Matrix"] == category]
    plt.scatter(subset["MDS1"], subset["MDS2"], label=category, color=color, alpha=0.6, edgecolor='k', s=70)

# Add legend
plt.legend(title="Matrix Kategorien", loc='best', fontsize=10)
plt.title("Relative Position der Proben (MDS)", fontsize=14)
plt.xlabel("MDS1", fontsize=12)
plt.ylabel("MDS2", fontsize=12)
plt.grid(True, alpha=0.3)

plt.show()
```
